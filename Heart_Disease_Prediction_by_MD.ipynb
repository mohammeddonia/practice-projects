# Heart Disease Prediction

Based on 13 attributes chosed among 74, the prediction is just to classify if the person has a heart disease or not. You can see the details of the data here (https://www.kaggle.com/ronitf/heart-disease-uci)

```python
# Basic Libraries Import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
from matplotlib.cm import rainbow
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

Here we will be experimenting with 3 algorithms 
1. KNeighborsClassifier
2. DecisionTreeClassifier
3. RandomForestClassifier

```python
# Classification Algorithms Import
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv('dataset.csv')
```

# Correlations Between Features

```python
import seaborn as sns
#get correlations of each features in dataset
corrmat = df.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(20,20))
#plot heat map
g=sns.heatmap(df[top_corr_features].corr(),annot=True,cmap="RdYlGn")
```

We can use the features with positive correlations only and see how the model performs. The most correlated features with the targer is [cp, thalach, slope, restecg]. For the time being, we will take all the features in consideration

```python
fig = plt.figure(figsize = (15,20))
ax = fig.gca()
df.hist(ax = ax)
```

We can consider this dataset as balanced dataset since the target histogram above is showing somewhat close quantities of both classes.

```python
df.head()
```

```python
df = pd.get_dummies(df, columns = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
standardScaler = StandardScaler()
columns_to_scale = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
df[columns_to_scale] = standardScaler.fit_transform(df[columns_to_scale])
```

```python
df.head()
```

```python
y = df['target']
X = df.drop(['target'], axis=1)
```

# KNN Classifier

```python
from sklearn.model_selection import cross_val_score
knn_scores = []
for k in range(1,21):
    knn_classifier = KNeighborsClassifier(n_neighbors = k)
    score=cross_val_score(knn_classifier,X,y,cv=10)
    knn_scores.append(score.mean())
```

```python
plt.figure(figsize=(20,20))
plt.plot([k for k in range(1, 21)], knn_scores, color = 'red')
for i in range(1,21):
    plt.text(i, knn_scores[i-1], (i, knn_scores[i-1]))
plt.xticks([i for i in range(1, 21)])
plt.xlabel('Number of Neighbors (K)')
plt.ylabel('Scores')
plt.title('K Neighbors Classifier scores for different K values')
```

# Random Forests Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf= RandomForestClassifier(n_estimators=15)

score=cross_val_score(clf,X,y,cv=10)
score.mean()
```

# Decision Trees

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
score = cross_val_score(clf,X,y,cv=10)
score.mean()
```
